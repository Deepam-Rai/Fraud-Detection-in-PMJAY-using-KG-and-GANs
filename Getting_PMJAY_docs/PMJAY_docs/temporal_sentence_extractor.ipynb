```python
from tika import parser
#TODO for OCR docs

def extract_text_from_pdf(filePath):
    parsed_file = parser.from_file(filePath)
    fileContent = parsed_file['content']
    fileContent = '' if fileContent is None else fileContent.strip()
    return fileContent
```

```python
%%time

import os
import spacy
import re
from timexy import Timexy
import pandas as pd

nlp = spacy.load("en_core_web_sm")

# Optionally add config if varying from default values
config = {
    "kb_id_type": "timex3",  # possible values: 'timex3'(default), 'timestamp'; Knowledge Base ID
    "label": "timexy",       # default: 'timexy'
    "overwrite": False       # default: False
}
nlp.add_pipe("timexy", config=config, before="ner")

file_count =0
temporal_sentences = []
filenames = []

# Replace 'path_to_directory' with the path to the root directory of your files
for root, _, files in os.walk(root_folder):
    for filename in files:
        file_count +=1
        print(f'file #:{file_count} \t temporal sentences: {len(temporal_sentences)} \t File: {filename}')
        if filename.endswith('.pdf'): # Change this to the file extension you're using
            file_path = os.path.join(root, filename)
            text = extract_text_from_pdf(file_path)
            text = re.sub('\n+', ' ', text) # replace multiple newline characters with a single space
            text = re.sub(' +', ' ', text) # replace multiple spaces with a single space
            doc = nlp(text)
            # Find temporal sentences
            for s in doc.sents:
                if set(['timexy','TIME','DATE','CARDINAL','ORDINAL']).intersection([e.label_ for e in s.ents]):
                # if set(['timexy','TIME','DATE']).intersection([e.label_ for e in s.ents]):
                    temporal_sentences.append(s)
                    filenames.append(filename)

temporals = pd.DataFrame({'Sentences':temporal_sentences, 'Filename':filenames})
```
