# Get links from the google sheet
to a dataframe

```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)

worksheet = gc.open('PMJAY_all_docs').sheet1

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()
print(rows)

# Convert to a DataFrame and render.
import pandas as pd
links = pd.DataFrame.from_records(rows)
links
```


# Download the files

```python
import os
import requests
import pandas as pd

# Load the dataframe
df = links
new_count = 0
already_count = 0

# Iterate through the rows of the dataframe
for index, row in df.iterrows():
    url = row[0].strip()

    # Check if the URL is not empty
    if url.startswith('http'):
        # Create the directory hierarchy
        path = os.path.join('downloads', *url.split('/')[2:-1])
        os.makedirs(path, exist_ok=True)
        
        # Download the file
        filename = url.split('/')[-1]
        filepath = os.path.join(path, filename)
        if not os.path.exists(filepath):
            print(f'Trying url:\t {url}')
            response = requests.get(url)
            with open(filepath, 'wb') as f:
                f.write(response.content)
            new_count +=1
        else:
            print(f'Already downloaded:\t {url}')
            already_count +=1
        print(f'\t\tCompleted: {new_count + already_count}\n')

print(f'Already Present: {already_count}\t Newly Downloaded: {new_count}')
print(f'Total files: {already_count + new_count}')
```
